```python
# Google Drive 마운트
from google.colab import drive
drive.mount('/content/drive')

# 필요한 라이브러리 설치
!pip install ultralytics
!pip install opencv-python-headless

# 라벨 파일 확인 및 수정 함수
import os

label_dirs = [
    '/content/drive/My Drive/Colab Notebooks/unified_dataset/train/labels',
    '/content/drive/My Drive/Colab Notebooks/unified_dataset/valid/labels',
    '/content/drive/My Drive/Colab Notebooks/unified_dataset/test/labels'
]

def fix_label_file(label_file_path):
    with open(label_file_path, 'r') as file:
        lines = file.readlines()

    corrected_lines = []
    for line in lines:
        parts = line.strip().split()
        if len(parts) != 5:
            print(f"Incorrect format in {label_file_path}: {line}")
            continue

        class_index, x_center, y_center, width, height = parts
        corrected_line = f"{class_index} {x_center} {y_center} {width} {height}\n"
        corrected_lines.append(corrected_line)

    if corrected_lines:
        with open(label_file_path, 'w') as file:
            file.writelines(corrected_lines)
        print(f"Fixed {label_file_path}")

for label_dir in label_dirs:
    for label_file in os.listdir(label_dir):
        label_file_path = os.path.join(label_dir, label_file)
        fix_label_file(label_file_path)

# data.yaml 파일 경로 수정
data_yaml_path = '/content/drive/My Drive/Colab Notebooks/unified_dataset/data.yaml'

# data.yaml 파일 내용 수정
new_data_yaml_content = """
train: /content/drive/My Drive/Colab Notebooks/unified_dataset/train/images
val: /content/drive/My Drive/Colab Notebooks/unified_dataset/valid/images
test: /content/drive/My Drive/Colab Notebooks/unified_dataset/test/images

nc: 3
names: ['mask', 'glasses', 'cap']
"""

with open(data_yaml_path, 'w') as file:
    file.write(new_data_yaml_content)

# 모델 학습 및 검증 코드
from ultralytics import YOLO

# 모델 로드 (사전 학습된 모델 없이 시작)
model = YOLO()

# GPU가 활성화되었는지 확인
import torch
print("CUDA available: ", torch.cuda.is_available())

# 모델 재학습
model.train(data=data_yaml_path, epochs=10, imgsz=640, batch=16, device=0)  # 'device' 파라미터를 '0'으로 설정하여 GPU 사용

# 학습 결과 확인
metrics = model.val()
print(metrics)

# 모델 저장
model.save('/content/drive/My Drive/Colab Notebooks/best_retrained.pt')
```
